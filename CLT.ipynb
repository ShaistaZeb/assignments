A continuous random variable $Z$ is said to be a standard normal (standard Gaussian) random variable, shown as $Zâˆ¼N(0,1)$, if its PDF is given by $$f_Z(z) = \frac{1}{\sqrt{2 \pi}} \exp\left\{-\frac{z^2}{2}\right\},    \hspace{20pt} \textrm{for all } z \in \mathbb{R}.$$ if$Z \sim N(0,1)$ then $E(Z)=0$ and $var(Z)=1$ <br>
<hr>
<b>CDF of the standard normal</b>

To find the CDF of the standard normal distribution, we need to integrate the PDF function. In particular, we have $$F_Z(z)=\frac{1}{\sqrt{2 \pi}}  \int_{-\infty}^{z}\exp\left\{-\frac{u^2}{2}\right\} du.$$ This integral does not have a closed form solution. Nevertheless, because of the importance of the normal distribution, the values of $F_Z(z)$ have been tabulated and many calculators and software packages have this function. We usually denote the standard normal CDF by $\Phi$.  <br>
<hr>
<b>Normal random variables</b>

Now that we have seen the standard normal random variable, we can obtain any normal random variable by shifting and scaling a standard normal random variable. In particular, define $$X=\sigma Z+\mu, \hspace{20pt} \textrm{where }\sigma > 0.$$ then $$EX=\sigma EZ+\mu=\mu,$$ $$\textrm{Var}(X)=\sigma^2  \textrm{Var}(Z)=\sigma^2.$$We say that $X$ is a normal random variable with mean $\mu$ and variance $\sigma^2$. We write $X\sim N(\mu,\sigma^2)$.<br>
<hr>
<b>Closed-form analytical approximations of Cumulate Distribution Functions (CDF) for
sum of random variables for which closed-form expressions are frequently not available or easy to
manipulate, but thanks to Central Limit Theorem, we can approximate any distribution to normal distribution by shifting and scaling the parameters. </b>
<hr>
The normal distribution is widely used. Part of the appeal is that it is well behaved and mathematically tractable. However, the central limit theorem provides a theoretical basis for why it has wide applicability.

<b>The central limit theorem basically states that as the sample size (n) becomes large, the following occur:</b><br>
<ul>
<li>The sampling distribution of the mean becomes approximately normal regardless of the distribution of the original variable.</li>
<li>
The sampling distribution of the mean is centered at the population mean, $\mu$, of the original variable. In addition, the standard deviation of the sampling distribution of the mean approaches $\frac{\sigma}{\sqrt n}$ </li>
